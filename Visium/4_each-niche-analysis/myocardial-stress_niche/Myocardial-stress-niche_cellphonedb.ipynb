## Prepare functions

```python
def revise_labels(cpdbouts):
    
    for old, new in {
                        'FB4':'FB4_activated',
                        'vCM3':'vCM3_stressed',
                        'NC1':'NC1_glial',
                        'NC2':'NC2_glial_NGF+',
                        'NC3':'NC3_glial',
                        'NC4':'NC4_glial',
                        'NC5':'NC5_glial',
                        'NC6':'NC6_schwann',
                        'EC7_atria':'EC7_endocardial'
                    }.items():
        cpdbouts.columns = [x.replace(old,new) for x in cpdbouts.columns]
    
    return cpdbouts
```

## Plot

```python
path_to_cpdb = '/nfs/team205/heart/cellphonedb/SP_onlymicroenv_Micropatho_0.1'

cpdbout_means = revise_labels(pd.read_table(f'{path_to_cpdb}/means.txt',index_col=0))
cpdbout_pvals = revise_labels(pd.read_table(f'{path_to_cpdb}/pvalues.txt',index_col=0))

post_cpdb.plot_CCI(cpdbout_means = cpdbout_means,
         target_celltype = 'FB4_activated',
         ligand_or_receptor = 'receptor',
         cpdbout_pvals = cpdbout_pvals,
         pvelue_thresh = 0.05,
         mean_thresh = 0.5,
         category_column = ['is_cytokine-associated'],
         interaction_metadata = pd.read_csv('/nfs/team205/heart/custom_cellphone/interaction_meta/interaction_metadata.csv'),
         microenvcells_to_plot = ['vCM3_stressed','FB3','FB4_activated','FB5','EC2_cap','EC3_cap','EC4_immune','EC6_ven',
                                  'PC3_str','SMC2_art','LYVE1+IGF1+MP','MoMP','NK_CD16hi'],
         fig_width=5.5,
             fig_hight=4,
          title='Cytokines',
         title_fontsize=20,
         title_hight=1.02
        )

plt.savefig('/home/jovyan/projects/P46_Adult-heart_Visium/notebooks/ver8regions_torepo/figures/CCI_SP_Cytokines-toFB4.pdf',
           bbox_inches='tight')
```
